```python
dataset = tf.data.Dataset.from_tensor_slices(list(join1))
```

```python
n_steps = 50
window_size = n_steps + 1
windows = dataset.window(window_size, shift = 1, drop_remainder =True).flat_map(lambda win:win.batch(window_size))
```

```python
for batch in windows.take(1):
    print(batch)
```

```python
keys   = tf.constant(["A","C","G","T","N"])
values = tf.constant([0,1,2,3,0], dtype=tf.int32)

table = tf.lookup.StaticHashTable(
    initializer=tf.lookup.KeyValueTensorInitializer(keys, values),
    default_value=4
)
```

```python
hash = windows.map(lambda w: table.lookup(w))
```

```python
def cou(da):
    count = 0
    for _ in da:
          count +=1
    return count
```

```python
dat_size = cou(hash)
```

```python
dat_size
```

```python
hash_shu = hash.shuffle(buffer_size=70000)
```

```python
train_size = int(0.80*(dat_size))

val_size = int(0.10*(dat_size))

test_size = dat_size - (train_size + val_size)
```

```python
train_size
```

```python
val_size
```

```python
test_size
```

```python
x_train = hash_shu.take(train_size)
x_val  = hash_shu.skip(train_size).take(val_size)
x_test = hash_shu.skip(train_size + val_size)
```

```python
x_train = x_train.batch(32)
x_val = x_val.batch(32)
x_test = x_test.batch(32)
```

```python
for i in x_train.take(1):
        print(i)
```

```python
x_trainoh = x_train.map(lambda i: tf.one_hot(i,depth = 4))
x_valoh = x_val.map(lambda i: tf.one_hot(i,depth = 4))
x_testoh = x_test.map(lambda i: tf.one_hot(i,depth = 4))
```
